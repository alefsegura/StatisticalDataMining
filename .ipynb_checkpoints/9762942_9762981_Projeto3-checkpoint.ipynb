```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import numpy as np
import pandas as pd
import seaborn as sns

import re

from sklearn.datasets import make_circles, load_diabetes
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold, cross_validate, StratifiedKFold
from sklearn.metrics import roc_auc_score, f1_score, make_scorer, roc_curve, auc
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, chi2, RFE
```

## Questão 1
<b>Considere	a	base	dados	gerado	pela	função	skdata.make_circles (noise	=	0.2), conforme	feito	em	aula. Compare	os	seguintes	algoritmos:</b>
- Redes	Neurais
- Naive	Bayes 
- Knn
- Random	forest

<b>Use	as	métricas	Accuracy, F1, Precision,	Recall. Use o	método k-fold	cross	validation	com	k=10.

```python
dataset = make_circles(1500, noise=0.2)

X = dataset[0]
y = dataset[1]

plt.scatter(X[:,0], X[:,1], c=y)
plt.title('Dataset')
plt.show()
```

```python
models = [
    MLPClassifier(max_iter=2000),
    KNeighborsClassifier(),
    GaussianNB(),
    RandomForestClassifier()
]

metrics = ['accuracy','f1','precision','recall']

kf = KFold(10)
scores = []
for model in models:
    for metric in metrics:
        cv = cross_validate(model, X, y, cv=kf, return_train_score=True, scoring=metric)
        score = {}
        score['Metric'] = metric
        score['Model'] = re.sub('Classifier','',type(model).__name__)
        score['Test Score'] = np.mean(cv['test_score'])
        score['Train Score'] = np.mean(cv['train_score'])
        scores.append(score)
df_scores = pd.DataFrame(scores)
            
# Resultados
plt.figure(figsize=(10,6))
plt.title('Desempenho dos classificadores por métrica')
sns.barplot(x=df_scores.Model, y=df_scores['Test Score'], hue=df_scores.Metric)
plt.show()
```

## Questão 2
<b>Considere a	partir	de	agora a	base	de	dados	sobre	<a href=https://www.kaggle.com/uciml/pima-indians-diabetes-database>diabetes</a>. Verifique	como a	classifição	muda	quando	consideramos	o	método k-fold cross	validation.	Ou	seja,	faça	um	gráfico	da	medida	Precison	em	função	de	k para	os	algoritmos	Redes	Neurais	e	Random	Forest.

```python
diabetes = load_diabetes()

X = diabetes['data']
y = [0 if c>=268 and c<=768 else 1 for c in diabetes['target']]

models = [
    MLPClassifier(max_iter=2000),
    KNeighborsClassifier(),
    GaussianNB(),
    RandomForestClassifier()
]

scores = []

for k in np.linspace(2,11,10,dtype=int):
    kf = KFold(k)
    for model in models:
        cv = cross_validate(model, X, y, cv=kf, return_train_score=True, scoring='precision', n_jobs=-1)
        score = {}
        score['Folds'] = k
        score['Model'] = re.sub('Classifier','',type(model).__name__)
        score['Test Score'] = np.mean(cv['test_score'])
        scores.append(score)

df_scores = pd.DataFrame(scores)
plt.figure(figsize=(12,6))
plt.title('Desempenho do classificador em função do número de Folds', size=18)
sns.lineplot(x=df_scores.Folds, y=df_scores['Test Score'], hue=df_scores.Model)
plt.xlabel('Número de folds', size=15)
plt.ylabel('Score',size=15)
plt.show()
```

## Questão 4
<b>Usando	o	método	que	seleciona	apenas	um	atributo	por	vez	(univariado, teste	qui-quadrado),	faça	uma	curva	da	medida	F1	em	função	do	número	de	atributos	considerados.	Isto	é,	obtenha	F1 para	o	atributo	mais	importante,	depois	para	os	dois	mais	importantes	e	assim	por	diante.

## Questão 5
Projete	os	dados	usando	PCA	e	faça	um	gráfico	de	F1	em	termos	do	número	de	components.	Quantos	components oferecem	a	maior	taxa	de	acerto?	Compare	com	a	curva	“cumulative	explained	variance”	versus	“number	of components”	e explique	os	resultados.

```python
pca = PCA(8)
X_pca = pca.fit_transform(X) 
X_train, X_test, y_train, y_test = train_test_split(X_pca,y,random_state=0)

componentes = []
scores = []

for i in range(len(pca.explained_variance_ratio_)):
    componentes.append({'acumulado':np.sum(pca.explained_variance_ratio_[:i+1]),
                        'variancia':pca.explained_variance_ratio_[i],
                        'n_components':i+1})
    for model in models:
        model.fit(X_train[:,:i+1],y_train)
        y_pred = model.predict(X_test[:,:i+1])
        score = {}
        score['Model'] = re.sub('Classifier','',type(model).__name__)
        score['F1'] = f1_score(y_test, y_pred)
        score['n_components'] = i+1
        scores.append(score)

df_comp = pd.DataFrame(componentes)
df_scores = pd.DataFrame(scores)

plt.figure(figsize=(8,10))

plt.subplot(2,1,1)
ax = sns.lineplot(x=df_scores.n_components, y=df_scores.F1, hue=df_scores.Model)
plt.title('Desempenho dos classificadores em função do número de componentes utilizados', size=17)
plt.xlabel('')
plt.ylabel('F1', size=14)

plt.subplot(2,1,2)
plt.title('Variância dos componentes', size=18)
sns.lineplot(x=df_comp.n_components, y=df_comp.acumulado)
sns.lineplot(x=df_comp.n_components, y=df_comp.variancia)
plt.ylabel('Aariância', size=14)
plt.xlabel('Número de componentes', size=14)
plt.legend(['Acumulada','Individual'])
plt.show()
```

<b>É possível	melhorar	a	classificação	considerando	um	número	reduzido	de atributos?

```python
titanic = pd.read_csv('titanic.csv')

X = titanic.drop(columns=['Survived'])
y = titanic.Survived

X['Sex'] = [0 if x=='male' else 1 for x in X.Sex]
X['embarked_C'] = [1 if c=='C' else 0 for c in titanic.Embarked]
X['embarked_S'] = [1 if c=='S' else 0 for c in titanic.Embarked]
X['embarked_Q'] = [1 if c=='Q' else 0 for c in titanic.Embarked]
X.drop(columns=['PassengerId','Name','Cabin','Ticket','Embarked'],inplace=True)
X.fillna(X.mean(), inplace=True)

models = [
    RandomForestClassifier(),
    MLPClassifier(max_iter=2000)
]

scores = []

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

plt.figure(figsize=(7,13))
scores = []
i = 1
for model in models:
    model.fit(X_train,y_train)
    y_proba = model.predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr,tpr)
    
    plt.subplot(2,1,i)
    plt.step(fpr, tpr, alpha=0.1)
    plt.fill_between(fpr, tpr, step='post', alpha=0.2, color='b')
    plt.text(0.4, 0.2, 'ROC area = %0.2f' % roc_auc, fontsize=13)
    plt.title('{} ROC Curve'.format(re.sub('Classifier','',type(model).__name__)), size=18)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', size=14)
    plt.ylabel('True Positive Rate', size=14)
    plt.xticks(size=10)
    plt.yticks(size=10)
    
    i += 1
    
plt.show()
```
