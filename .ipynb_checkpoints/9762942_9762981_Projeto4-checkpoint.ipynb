```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import scipy.cluster.hierarchy as sch
import numpy as np
import pandas as pd
import seaborn as sns
import re

import sklearn.datasets as skdata
from sklearn.cluster import KMeans, SpectralClustering, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, normalized_mutual_info_score
from sklearn.decomposition import PCA
```


## Questão 1
<b>Usando	o	método	k-means,	determine	o	número	ideal	de	sementes	(Elbow	Method)	para	noise=0.1,	0.2	e	0.5.

```python
n_samples=200
i = 1
plt.figure(figsize=(16,4))
for n in [0.1, 0.2, 0.5]:
    X, y = skdata.make_moons(n_samples=n_samples, noise=n)
    d = np.column_stack((X, np.transpose(y)))
    data = pd.DataFrame(data=d, columns=['X1','X2','y'])
    features_names = ['X1','X2']
    class_labels = np.unique(y)

    #Plotting
    plt.subplot(1,3,i)
    colors = ['red','blue','green','black']
    aux = 0
    for c in class_labels:
        ind = np.where(y==c)
        plt.title('Dados com %.1f de ruído' % n)
        plt.scatter(X[ind,0][0], X[ind,1][0], color=colors[aux], label=c)
        aux = aux + 1
    i += 1
plt.show()
```

```python
N = [0.1, 0.2, 0.5]
K = np.arange(1,20,1)

df = []
for n in N:
    result = []
    X, y = skdata.make_moons(n_samples=n_samples, noise=n)
    for k in K:
        km = KMeans(k).fit(X)
        result.append({'K':k,n:km.inertia_})
    df.append(pd.DataFrame(result).set_index('K'))

df = pd.concat(df, axis=1)
df.plot(figsize=(12,5))
plt.title('Desempenho do K-means em função de K para diferentes níveis de ruído', size=18)
plt.ylabel('Sum of squared errors', size=13)
plt.xlabel('K', size=13)
plt.show()
```

## Questão 2
<b>Construa	um	gráfico	da	medida	silhouette_score	em	função	do	nível	de	ruído	noise.	Considere	os	métodos	k-means	e	spectral	clustering.

```python
models = [KMeans(), SpectralClustering()]

result = []
for n in N:
    X, y = skdata.make_moons(n_samples=n_samples, noise=n)
    for mod in models:
        m = mod.fit(X)
        score = silhouette_score(X,m.labels_)
        result.append({'model':type(m).__name__,'N':n,'Score':score})

df = pd.DataFrame(result)
plt.figure(figsize=(9,6))
sns.barplot(df['N'], df['Score'], hue=df['model'])
plt.title('Desempenho dos modelos K-means e SpectralClustering em função do nível de ruído')
plt.xlabel('Noise', size=15)
plt.ylabel('Score', size=15)
plt.show()
```

## Questão 3
<b>Para	o	método	DBSCAN,	varie	os	valores	eps	e	min_samples	para	obter	o	máximo	da	medida	silhouette_score.	Grafique	os	resultados	para	os	melhores	valores	encontrados.

```python
X,_ = skdata.make_moons(n_samples=n_samples, noise=0.2)

scores = np.zeros((3,3))
eps = [0.1, 0.15, 0.2]
mins = [3,4,5]
for i in range(len(eps)):
    for j in range(len(mins)):
        dbs = DBSCAN(eps=eps[i], min_samples=mins[j], n_jobs=-1).fit(X)
        scores[i,j] = silhouette_score(X, dbs.labels_)        

plt.figure(figsize=(8,5))
sns.heatmap(scores, annot=True, xticklabels=mins, yticklabels=eps, cmap='Blues')
plt.title('Silhouette Score do DBSCAN em função dos parâmetros', size=16)
plt.xlabel('min_samples', size=14)
plt.ylabel('eps', size=14)
plt.show()
```

## Questão 4
<b>Considere	os	dados	da	Iris.	Construa	os	dendrogramas	usando	Agglomerative	Hierarchical	Clustering	com	os	métodos	average	linkage,	Ward	e	Complete	Linkage.	Encontre	o	melhor	valor	de	k	e	mostre	os	dados	projetados	usando	PCA	e	os	clusters	obtidos.

```python
dataset = skdata.load_iris()

X = dataset['data']
y = dataset['target']

X2 = PCA(2).fit_transform(X)
```

```python
i=1
y_hc = []
for method in ['average', 'ward','complete']:
    plt.figure(figsize=(10,5))
    plt.title(method.upper()+' LINKAGE', size=22)
    dendrogram = sch.dendrogram(sch.linkage(X, method=method))
    hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage=method)
    plt.xticks(size=8)
    plt.yticks(size=13)
    plt.show(True)
    y_hc.append(hc.fit_predict(X))
```

```python
i=1
plt.figure(figsize=(18,5))
for y,t in zip(y_hc,['average', 'ward','complete']):
    plt.subplot(1,3,i)
    plt.title(t.upper()+' LINKAGE', size=16)
    plt.scatter(X2[:,0],X2[:,1],c=y)
    i+=1
plt.show()
```

## Questão 5
<b>Compare	os	métodos	de	clustering	para	os	dados	da	iris	usando	a	medida	<a href=https://scikitlearn.org/stable/modules/generated/sklearn.metrics.normalized_mutual_info_score.html>normalize	mutual	information<a/>

```python
dataset = skdata.load_iris()

X = dataset['data']
y = dataset['target']

models = [KMeans(),
          SpectralClustering(),
          DBSCAN()]

scores = []

for model in models:
    model.fit(X)
    score = normalized_mutual_info_score(y, model.labels_)
    scores.append({'model':re.sub('Classifier','',type(model).__name__),'score':score})
    
df_scores = pd.DataFrame(scores)

plt.figure(figsize=(7,5))
plt.bar(df_scores.model, df_scores.score)
plt.ylabel('Normalized Mutual Information Score', size=13)
plt.xlabel('Model', size=13)
plt.title('Desempenho dos Agrupadores', size=20)
plt.show()
```
